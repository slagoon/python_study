```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df_2011 = df[df['Year']>=2011].reset_index(drop=True)
pivot_df = pd.pivot_table(df_2011, index='Year', columns='Country',
                            values = 'Life_Expectancy')
pivot_df
```
